```python
import gradio as gr

from transformers import TFAutoModelForSequenceClassification, AutoTokenizer
import numpy as np
```

```python
# load model
model_v1 = TFAutoModelForSequenceClassification.from_pretrained('drive/MyDrive/FIRE/OOD/train_on_sst2/bert')
base_model = "distilbert-base-uncased"
tokenizer_v1 = AutoTokenizer.from_pretrained(base_model)
```

```python
def model_inference(sen):
    tokenized_v1 = tokenizer_v1([sen], return_tensors="np", padding="longest")
    outputs_v1 = model_v1(tokenized_v1).logits
    classifications_v1 = np.argmax(outputs_v1, axis=1)
    if classifications_v1[0] == 1:
        res = "Positive :)"
    else:
        res = "Negative :("
    return res
```

```python
demo = gr.Interface(
    fn=model_inference,
    inputs=gr.Textbox(lines=2, placeholder="Enter a Sentence"),
    outputs="text",
)
demo.launch()
```
